# 12장 분류

```python
# 필요한 라이브러리를 임포트합니다.
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

```

```python
# 아이리스 데이터셋을 로드합니다.
iris = load_iris()

# 데이터를 훈련 세트와 테스트 세트로 나눕니다 (훈련 세트: 80%, 테스트 세트: 20%).
trn_data, test_data, trn_target, test_target = train_test_split(iris.data, iris.target, train_size=0.8)
```

```python
# KNN 분류기를 생성하고 학습시킵니다 (이웃의 수: 3).
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(trn_data, trn_target)

# 테스트 데이터에 대한 예측을 수행합니다.
clf.predict(test_data)

# 테스트 데이터에 대한 정확도를 계산합니다.
print("Iris KNN Score:", clf.score(test_data, test_target))
```

```python
# 와인 데이터셋을 로드합니다.
from sklearn.datasets import load_wine
wine = load_wine()
data = wine.data
target = wine.target

# 데이터를 훈련 세트와 테스트 세트로 나눕니다 (훈련 세트: 80%, 테스트 세트: 20%).
trn_data, test_data, trn_target, test_target = train_test_split(data, target, train_size=0.8)
```

```python
# KNN 분류기를 생성하고 학습시킵니다 (이웃의 수: 5).
clf = KNeighborsClassifier(n_neighbors=5)
clf.fit(trn_data, trn_target)

# 테스트 데이터에 대한 예측을 수행합니다.
pred = clf.predict(test_data)

# 테스트 데이터에 대한 정확도를 계산합니다.
score = clf.score(test_data, test_target)
print("Wine KNN Score:", score)
```

```python
# KNN 분류기의 평균 정확도를 계산하기 위해 1000번 반복합니다.
avgscore = 0
for run in range(1000):
    trn_data, test_data, trn_target, test_target = train_test_split(data, target, train_size=0.8)
    clf = KNeighborsClassifier(n_neighbors=5)
    clf.fit(trn_data, trn_target)
    pred = clf.predict(test_data)
    score = clf.score(test_data, test_target)
    print(f"run={run}, score={score}")
    avgscore += score
print(f"Average KNN Score over 1000 runs: {avgscore / 1000}")
```

# 결정 트리 분류기를 사용하여 와인 데이터셋을 학습하고 평가


```python
# 결정 트리 분류기를 사용하여 와인 데이터셋을 학습하고 평가합니다.
from sklearn.tree import DecisionTreeClassifier
wine = load_wine()
data = wine.data
target = wine.target
```

```python
# 데이터를 훈련 세트와 테스트 세트로 나눕니다 (훈련 세트: 80%, 테스트 세트: 20%).
trn_data, test_data, trn_target, test_target = train_test_split(data, target, train_size=0.8)
```

```python
# 결정 트리 분류기를 생성하고 학습시킵니다.
clf = DecisionTreeClassifier()
clf.fit(trn_data, trn_target)

# 테스트 데이터에 대한 예측을 수행합니다.
pred = clf.predict(test_data)

# 테스트 데이터에 대한 정확도를 계산합니다.
score = clf.score(test_data, test_target)
print("Wine Decision Tree Score:", score)
```

```python
# 결정 트리 분류기의 평균 정확도를 계산하기 위해 1000번 반복합니다.
avgscore = 0
for run in range(1000):
    trn_data, test_data, trn_target, test_target = train_test_split(data, target, train_size=0.8)
    clf = DecisionTreeClassifier()
    clf.fit(trn_data, trn_target)
    pred = clf.predict(test_data)
    score = clf.score(test_data, test_target)
    print(f"run={run} score={score}")
    avgscore += score
print(f"Average Decision Tree Score over 1000 runs: {avgscore / 1000}")
```

```python
# 결정 트리 모델을 시각화합니다.
from sklearn.tree import export_graphviz
import graphviz
```

```python
#결정 트리의 구조를 그래프로 생성합니다.
dot_data = export_graphviz(clf, out_file=None,
                           feature_names=wine.feature_names,
                           class_names=wine.target_names,
                           filled=True, rounded=True,
                           special_characters=True)
graph = graphviz.Source(dot_data)
graph.render("wine")

# 그래프를 출력합니다.
graph
```
