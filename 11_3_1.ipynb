```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import numpy as np

# 아이리스 데이터셋 불러오기
iris = load_iris()

# KMeans 모델 생성 및 데이터 학습
km = KMeans(n_clusters=3, random_state=15)  # KMeans 모델 생성, 클러스터 개수는 3으로 설정
km.fit(iris.data)  # 모델을 데이터에 적합시킴

# 클러스터 예측
labels = km.predict(iris.data)  # 데이터를 클러스터에 할당

# 클러스터 예측 결과 출력
print(labels)

# 클러스터별 정확도 확인
correct, incorrect = 0, 0
for i in range(0, 50):
    if labels[i] == 0:  # Setosa 종류는 label이 0으로 예상
        correct += 1
    else:
        incorrect += 1
print("setosa \t correct:", correct, " incorrect:", incorrect)

correct, incorrect = 0, 0
for i in range(50, 100):
    if labels[i] == 1:  # Versicolor 종류는 label이 1으로 예상
        correct += 1
    else:
        incorrect += 1
print("versicolor \t correct:", correct, " incorrect:", incorrect)

correct, incorrect = 0, 0
for i in range(100, 150):
    if labels[i] == 2:  # Virginica 종류는 label이 2으로 예상
        correct += 1
    else:
        incorrect += 1
print("virginica \t correct:", correct, " incorrect:", incorrect)

# 아이리스 데이터셋의 일부를 산점도로 시각화
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=labels)  # 첫 두 개의 피처를 사용하여 산점도 그리기, 클러스터 레이블을 색상으로 표시
plt.xlabel('sepal length (cm)')  # x축 라벨
plt.ylabel('sepal width (cm)')  # y축 라벨
plt.title('Iris Data - Sepal Length vs Width')  # 그래프 제목
plt.show()

# 클러스터의 최적 개수 찾기 (엘보우 방법)
inertia = []  # 클러스터 내 제곱합을 저장할 리스트
for k in range(2, 8):  # 클러스터 개수를 2에서 7까지 변경하며 실험
    km = KMeans(n_clusters=k, random_state=15)  # k개의 클러스터를 가진 KMeans 모델 생성
    km.fit(iris.data)  # 모델을 데이터에 적합시킴
    inertia.append(km.inertia_)  # 클러스터 내 제곱합을 리스트에 추가

# 관성(inertia) 값을 시각화
plt.plot(range(2, 8), inertia)  # 클러스터 개수에 따른 관성 값 그래프
plt.xlabel('Number of clusters (k)')  # x축 라벨
plt.ylabel('Inertia')  # y축 라벨
plt.title('Elbow Method for Optimal k')  # 그래프 제목
plt.show()
```
