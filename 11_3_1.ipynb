```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import numpy as np

#   
iris = load_iris()

# KMeans     
km = KMeans(n_clusters=3, random_state=15)  # KMeans  ,   3 
km.fit(iris.data)  #   

#  
labels = km.predict(iris.data)  #   

#    
print(labels)

#   
correct, incorrect = 0, 0
for i in range(0, 50):
    if labels[i] == 0:  # Setosa  label 0 
        correct += 1
    else:
        incorrect += 1
print("setosa \t correct:", correct, " incorrect:", incorrect)

correct, incorrect = 0, 0
for i in range(50, 100):
    if labels[i] == 1:  # Versicolor  label 1 
        correct += 1
    else:
        incorrect += 1
print("versicolor \t correct:", correct, " incorrect:", incorrect)

correct, incorrect = 0, 0
for i in range(100, 150):
    if labels[i] == 2:  # Virginica  label 2 
        correct += 1
    else:
        incorrect += 1
print("virginica \t correct:", correct, " incorrect:", incorrect)

#     
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=labels)  #       ,    
plt.xlabel('sepal length (cm)')  # x 
plt.ylabel('sepal width (cm)')  # y 
plt.title('Iris Data - Sepal Length vs Width')  #  
plt.show()

#     ( )
inertia = []  #     
for k in range(2, 8):  #   2 7  
    km = KMeans(n_clusters=k, random_state=15)  # k   KMeans  
    km.fit(iris.data)  #   
    inertia.append(km.inertia_)  #     

# (inertia)  
plt.plot(range(2, 8), inertia)  #      
plt.xlabel('Number of clusters (k)')  # x 
plt.ylabel('Inertia')  # y 
plt.title('Elbow Method for Optimal k')  #  
plt.show()
```
